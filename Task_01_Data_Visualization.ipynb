```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (CORRECT PATH)
df = pd.read_csv(
    "/API_SP.POP.TOTL_DS2_en_csv_v2_174326.csv",
    skiprows=4
)

# Preview data
df.head()

# Find year columns
year_columns = [col for col in df.columns if col.isdigit()]
latest_year = year_columns[-1]

# Convert to numeric
population = pd.to_numeric(df[latest_year], errors="coerce")

# Plot histogram
plt.figure(figsize=(8,5))
plt.hist(population.dropna(), bins=20)
plt.title(f"Population Distribution Across Countries ({latest_year})")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.show()

```
